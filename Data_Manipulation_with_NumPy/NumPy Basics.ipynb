```python
import numpy as np

# arange

arr = np.arange(5)
print(repr(arr))

arr = np.arange(5.1)
print(repr(arr))

arr = np.arange(-1, 4)
print(repr(arr))

arr = np.arange(-1.5, 4, 2)
print(repr(arr))

# linspace

# The end of the range is inclusive for np.linspace, unless the keyword argument endpoint is set to False.
# To specify the number of elements, we set the num keyword argument (its default value is 50).

# default number of elements is 50
arr = np.linspace(5, 11)
print(repr(arr))

arr = np.linspace(5, 11, num=4)
print(repr(arr))

arr = np.linspace(5, 11, num=4, endpoint=False)
print(repr(arr))

arr = np.linspace(5, 11, num=4, dtype=np.int32)
print(repr(arr))

arr = np.arange(8)

reshape_arr = np.reshape(arr, (2,4))
print(repr(reshape_arr))

reshape_arr = np.reshape(arr, (-1, 2, 2))
print(repr(reshape_arr))
print('New Shape: {}'.format(reshape_arr.shape))

# faltten()

arr = np.arange(8).reshape((2,4))
flatten = arr.flatten()
print(repr(arr))
print(repr(flatten))

# transpose(arr)

arr = np.arange(8).reshape((4,2))
transposed = np.transpose(arr)
print(repr(arr))
print(repr(transposed))

arr = np.arange(24).reshape((3,4,2))
transposed = np.transpose(arr, axes=(1,2,0))
print(repr(arr))
print('arr shape: {}'.format(arr.shape))
print('transpose shape: {}'.format(transposed.shape))
```
